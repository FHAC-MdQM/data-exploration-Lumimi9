# Loading and exploring spring parking brake data

## Systembeschreibung

Das zu betrachtende System ist ein Federspeicherbremszylinder, also ein Bremszylinder, der ohne angelegten Luftdruck die Bremse anlegt. Abhängig von der Federspeicherkraft kann das Fahrzeug (bspw. eine Lokomotive mit vier Federspeichern pro Fahrzeug) im Gefälle stationär gehalten werden oder nicht.

Die 24 Federn pro Zylinder verhalten sich näherungsweise linear, weisen jedoch eine recht große Streuung der Federrate auf. Sie sind zusammen mit einigen geometrischen Größen für die Kraft verantwortlich:

- Force/N: gemessene Kraft des Federspeichers
- Deviation $d$/mm: Abweichung der Dicke des Kolben (Abweichung nach unten: weniger Kraft)
- Deviation $h_{1}$/mm: Abweichung der Lage der Bohrung des Hammerheads (Abweichung nach unten: weniger Kraft)
- Deviation $h_{2}$/mm: Abweichung der Lage der Bohrung des Hammerheads (Abweichung nach unten: weniger Kraft)
- Dev. $l$/mm: Abweichung der Länge der Kolbenstange (Abweichung nach unten: weniger Kraft)

Zusätzlich haben die Daten noch Datum und Zeit der Messung.

## Aufgabe

## Library import

```python
# OS to use file system
import os
# Matplotlib and the pyplot module for a Matlab-like plot syntax
import matplotlib
# This is the syntax to import as a named module
import matplotlib.pyplot as plt
# Pandas is a module for rectangular data "dataframe"
import pandas as pd
# Special module for a very quick analysis
from pandas.plotting import scatter_matrix
# Plot the graphs in the notebook
%matplotlib inline
```

## Data import

```python
# Import the csv-data to pandas dataframe (canonical name df)
df = pd.read_csv(os.path.join("SpringPBData.csv"))
# Display the first lines of the df
df.head()
```

```python
# Transform first column from string to datetimeindex
df.set_index(pd.to_datetime(df["Time"]), inplace = True)
# Delete second ocurrence of time
df.drop("Time", axis = 1, inplace = True)
df.head()
```

```python
df.describe()
```

# Auswertung

Im ersten Schritt der Auwertung wird eine Scatter Matrix dargestellt, diese ermöglicht einen schnellen und leichten überblick über die einzelnen zusammenhänge der Variablen. 

```python
q = scatter_matrix(df)
```

Die Plots lassen eine Lineare Abhängigkeit der Kraft von der Lage der Bohrungen des Hammerheads (h1 und h2), sowie die Länge der Kolbenstange (l) vermuten.

Im weiteren scheinen die Größen h1 und h2 linear voneinander abzuhängen. 

Die Restlichen Kombinationen deuten nicht auf Korrelationen hin. 

## Einflussgrößen - Plots

Im Folgenden wird der zusammenhang zwischen der Kraft und h1, h2 sowie l untersucht. 

```python
print("h1-F Diagramm")
df.plot.scatter(y="F", x="h1")
```

Auch der Zusammenhang zwischen h1 und der Kraft ist von starken 
Streuungen betroffen.
Ein Grenzwert für h1 der bei Unerschreitung direkt zum unterschreiten der Grenzkraft F führt kann nicht eindeutig festgestellt werden. 

```python
print("h2-F Diagramm")
df.plot.scatter(y="F", x="h2")
```

Der Zusammenhang zwischen F und h2 unterliegt deutlich geringerer Streuung. Ein unterschreiten des wertes von h2 etwa unter h2=-0.8 führt bei den aufgenommenen Federspeichern zum unterschreiten der Kritischen Kraft. 

```python
print("F-l Diagramm")
df.plot.scatter(y="F", x="l")
```

Der Zusammenhang zwischen F und l unterliegt ebenso wie d und h1 starker Streuung, so gibt es auch hier keinen schwellenwert, der zu einem Eindeutigen über oder unterschreiten der notwendigen Federkraft führt. 

```python
print("h2-F Diagramm")
print("h2-F Diagramm")
df.plot.scatter(y="d", x="h2", c="F", cmap="Greens")
df.plot.scatter(y="h1", x="h2", c="F", cmap="Greens")
df.plot.scatter(y="l", x="h2", c="F", cmap="Greens")
```

Zur weiteren bestärkung des Zusammenhanges zwischen der Kraft F und h2 lässt sich in den plots F2 gegen d, h1 und l gut erkennen, das der Farbverlauf ausnahmslos in x-richtung also mit h2 ansteigt. Somit ist eindeutig das hier der Zusammenhang zwischen h2 und F überwiegt.
